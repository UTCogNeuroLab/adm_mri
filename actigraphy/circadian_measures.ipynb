# Calculate cosinor measures for ADM study

## Preprocessing

```python
import os
import glob
import numpy as np
import pandas as pd
```

```python
sr = '30s' # 1 minute, downsampling Actiwatch from 30 s to 1 min
recording_period_min = [7]
interpolate_limit = 10 # 10 minute limit
```

```python
home_dir = os.path.expanduser("~")
data_dir = os.path.join(home_dir, 'Box/CogNeuroLab/Aging Decision Making R01/data/actigraphy/raw/')

act_files = glob.glob(data_dir + "*New_Analysis.csv")
```

```python
act_files
```

```python
act = {}
```

```python
from wearables import preproc

out_dir = os.path.join(data_dir, 'processed_2020-08-18')

for in_file in act_files:
    print(in_file)
    record_id = str.split(in_file, "raw/")[-1][0:5]
    act[record_id] = {}

    for ndays in recording_period_min:
        if in_file.endswith('.csv'):
            act[record_id][str(ndays)] = preproc.preproc(in_file, out_dir, 'actiwatch', sr, ndays, interpolate_limit, truncate=True, write=False)
```

```python
import numpy as np
np.save(home_dir + '/Box/CogNeuroLab/Aging Decision Making R01/data/actigraphy/processed_2020-08-18/act_preproc.npy', act)
```

```python
import matplotlib.pyplot as plt
plt.plot(act['40891']['7'].index, act['40891']['7'])
```

## Cosinor Analysis

Load saved npy files to save time.

```python
act = np.load(home_dir + '/Box/CogNeuroLab/Aging Decision Making R01/data/actigraphy/processed_2020-08-18/act_preproc.npy', allow_pickle=True)[()]
```

```python
act['40891']
```

```python
cr_measures = ['actmin', 'amp', 'alpha', 'beta', 'phi']
cols = [col + "_" + str(ndays) for ndays in recording_period_min for col in cr_measures]
```

```python
[str.split(in_file, "raw/")[-1][0:5] for in_file in act_files]
```

```python
from wearables import fitcosinor

cr_act = pd.DataFrame(index = [str.split(in_file, "raw/")[-1][0:5] for in_file in act_files], columns = cols)

for record_id in act.keys():
    print(record_id)
    for ndays in act[record_id].keys():
        try:
            cr_act.loc[record_id]["actmin_" + str(ndays) : "phi_" + str(ndays)] = fitcosinor.fitcosinor(act[record_id][str(ndays)], transform = 'antilogistic')[0].values[0]
        except Exception as e:
            print(e)
```

```python
cr_act.to_csv(home_dir + '/Box/CogNeuroLab/Aging Decision Making R01/data/actigraphy/circadian_measures/7_days/cr_act_py.csv')
```
